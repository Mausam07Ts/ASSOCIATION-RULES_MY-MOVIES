```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import numpy as np
```

```python
df_mov = pd.read_csv("my_movies.csv")
```

```python
df_mov
```

```python
df_mov.shape
```

```python
df = df_mov.iloc[:,5:]
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.9)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
rules[rules.lift>5]#FROM BELOW WE SEE THAT IF PERSON BUYS LOTR THEN HE WILL LIKELY BUY GLADIATOR AND GREEN MILE AND VICE VERSA
```

```python
frequent_itemsets2 = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets2#CHANG MIN_SUPPORT TO 0.5
```

```python
rules2 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.8)
rules2
```
